```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Load ticker and post data
tickers = pd.read_csv('../../data/tabular/tickers.csv')

posts = pd.read_csv('../../data/tabular/wsb_posts/cleaned posts.csv',
                    index_col=0)
posts = posts.drop(columns=[col for col in posts.columns if 'Unnamed' in col])

posts['date'] = pd.to_datetime(posts['created_utc']).dt.date
posts = posts.drop(columns=['created_utc'])
```

```python
# calculate the number of unique tickers present in the posts data
unique_tickers = pd.Series(list(set(','.join(posts.tickers).split(','))), name='Ticker').sort_values()
n_unique = unique_tickers.shape[0]
print('There are {} unqiue tickers mentioned in WSB comments.'.format(n_unique))
```

```python
# Calculate the retail popularity for each ticker on each day. This will be a 
# n x m matrix where n is number of days and m is number of tickers present in the comments

rh_prefix = '../../data/tabular/popularity_export'
ticker = unique_tickers.iloc[0]

def aggregate_popularity(ticker):
    rh_prefix = '../../data/tabular/popularity_export'
    ticker_df = pd.read_csv('{}/{}.csv'.format(rh_prefix, ticker), 
                            parse_dates=[0])
    ticker_df = (
        ticker_df
        .groupby(ticker_df.timestamp.rename('date').dt.date)
        .last()
        [['users_holding']]
        .rename(columns={'users_holding': ticker})
    )
    return ticker_df


def load_popularities(tickers):
    popularity_df = aggregate_popularity(tickers.iloc[0])
    
    for idx, ticker in tickers[1:].iteritems():
        next_ticker   = aggregate_popularity(ticker)
        popularity_df = pd.merge(popularity_df,
                                 next_ticker,
                                 left_index=True, 
                                 right_index=True,
                                 how='left')
        
    return popularity_df

popularity_csv = '../../data/tabular/accounts_holding_table.csv'

if os.path.exists(popularity_csv):
    popularity_df = pd.read_csv(popularity_csv)
else:
    popularity_df = load_popularities(unique_tickers)
    popularity_df.to_csv('../../data/tabular/accounts_holding_table.csv')
```

```python
# unpack tickers for each comment
data_by_ticker = []

for idx, row in posts.iterrows():
    for ticker in row.tickers.split(','):
        data_by_ticker.append(
            {
                'num_comments': row.num_comments,
                'score': row.score,
                'tickers': ticker,
                'date': row.date
            }
        )
        
expanded_posts = pd.DataFrame(data_by_ticker)
```

```python
def plot_interaction_series(ticker, post_df):
    proc_df = (
        post_df
        .loc[post_df.tickers == ticker]
        .groupby('date')
        .sum()
    )
    
    fig, ax = plt.subplots(2, 1, figsize=(16, 8))
    ax[0].plot(proc_df.index, proc_df.score, color='blue', label='Aggregate Score')
    ax[0].legend()
    ax[1].plot(proc_df.index, proc_df.num_comments, color='gray', label='Number of Comments')
    ax[1].legend()
    plt.show()
    
plot_interaction_series('GOOG', expanded_posts)
```

```python
agg_posts = (
    expanded_posts
    .groupby(['date', 'tickers'])
    .sum()
    .apply(lambda x: np.log(1 + x))
)
sns.pairplot(agg_posts)
```

```python
# load returns and volume by tailors 
def load_market_ticker(ticker):
    prefix = "../../data/tabular/daily_price"
    market_df = pd.read_csv('{}/{}.csv'.format(prefix, ticker), 
                            parse_dates=['Date'])
    market_df['PriceReturn'] = market_df['Close'] / market_df['Close'].shift(1) - 1
    return (
        market_df
        [['Date', 'Volume', 'PriceReturn']]
        .dropna()
        .set_index('Date')
    )
    

def load_market(tickers):
    
    market_df = load_market_ticker(tickers.iloc[0])
    volume_df = market_df[['Volume']].rename(columns={'Volume': tickers.iloc[0]})
    return_df = market_df[['PriceReturn']].rename(columns={'PriceReturn': tickers.iloc[0]})
    
    for idx, ticker in tickers[1:].iteritems():
        next_market = load_market_ticker(ticker)
        next_volume = next_market[['Volume']].rename(columns={'Volume': ticker})
        next_return = next_market[['PriceReturn']].rename(columns={'PriceReturn': ticker})
    
        volume_df = pd.merge(volume_df, next_volume, left_index=True, right_index=True, how='left')
        return_df = pd.merge(return_df, next_return, left_index=True, right_index=True, how='left')
    
    return return_df, volume_df
return_df, volume_df = load_market(unique_tickers)
```

```python
return_df.to_csv('../../data/tabular/returns.csv')
volume_df.to_csv('../../data/tabular/volume.csv')
```

```python
depivot_volume_df = (
    volume_df
    .reset_index()
    .melt(id_vars='Date',
          var_name='Ticker',
          value_name='Volume')
)

depivot_return_df = (
    return_df
    .reset_index()
    .melt(id_vars='Date',
          var_name='Ticker',
          value_name='PriceReturn')
)
```

```python
def trailing_returns(df, ticker, days_trailing):
    return (
        (df
         .loc[depivot_return_df.Ticker==ticker,
              ['Date', 'PriceReturn']]
         .groupby('Date')
         .sum() + 1)
        .rolling(window=days_trailing)
        .apply(np.prod)
        .dropna()
    ) - 1

tickers = sorted(depivot_return_df.Ticker.unique())



quarter_trailing_df = (
    trailing_returns(depivot_return_df, tickers[0], 63)
    .rename(columns={'PriceReturn': tickers[0]})
)
month_trailing_df = (
    trailing_returns(depivot_return_df, tickers[0], 20)
    .rename(columns={'PriceReturn': tickers[0]})
)

for i, ticker in enumerate(tickers[1:]):
    last_ticker = tickers[i-1]
    quarter_trailing_df = (
        quarter_trailing_df.merge(
            trailing_returns(depivot_return_df, ticker, 63),
            left_index=True, right_index=True, how='left')
        .rename(columns={'PriceReturn': ticker})
    )
    month_trailing_df = (
        month_trailing_df.merge(
            trailing_returns(depivot_return_df, ticker, 20),
            left_index=True, right_index=True, how='left')
        .rename(columns={'PriceReturn': ticker})
    )
    
quarter_trailing_df.to_csv('../../data/tabular/quarter-trailing-returns.csv')
month_trailing_df.to_csv('../../data/tabular/month-trailing-returns.csv')
```

```python
annual_trailing_df = (
    trailing_returns(depivot_return_df, tickers[0], 252)
    .rename(columns={'PriceReturn': tickers[0]})
)

semiannual_trailing_df = (
    trailing_returns(depivot_return_df, tickers[0], 126)
    .rename(columns={'PriceReturn': tickers[0]})
)

for i, ticker in enumerate(tickers[1:]):
    last_ticker = tickers[i-1]
    annual_trailing_df = (
        annual_trailing_df.merge(
            trailing_returns(depivot_return_df, ticker, 252),
            left_index=True, right_index=True, how='left')
        .rename(columns={'PriceReturn': ticker})
    )
    semiannual_trailing_df = (
        semiannual_trailing_df.merge(
            trailing_returns(depivot_return_df, ticker, 126),
            left_index=True, right_index=True, how='left')
        .rename(columns={'PriceReturn': ticker})
    )
    
annual_trailing_df.to_csv('../../data/tabular/annual-trailing-returns.csv')
semiannual_trailing_df.to_csv('../../data/tabular/semiannual-trailing-returns.csv')
```

```python
# Calculate cross sectional momentum on quarterly and monthly trailing returns

def momentum_depivot(trailing_returns_df, popularity_df):
    returns_rank_df = trailing_returns_df.rank(axis=1)
    popularity_rank_df = popularity_df.rank(axis=1)
    
#     returns_rank_df.index = pd.to_datetime(returns_rank_df.index)
    popularity_rank_df.index = pd.to_datetime(popularity_rank_df.index)
    
    returns_rank_df.index.name = 'Date'
    popularity_rank_df.index.name = 'Date'
    
    depivot_returns = (
        returns_rank_df
        .reset_index()
        .melt(id_vars=['Date'],
              var_name='Ticker',
              value_name='Rank')
    )
    
    depivot_popularity = (
        popularity_rank_df
        .reset_index()
        .melt(id_vars=['Date'],
              var_name='Ticker',
              value_name='Rank')
    )
    
    depivot_df = pd.merge(
        depivot_returns,
        depivot_popularity,
        on = ('Date', 'Ticker'),
        suffixes = ('_mom', '_rh')
    )
    return depivot_df
```

```python
quarter_depivot    = momentum_depivot(quarter_trailing_df, popularity_df)
month_depivot      = momentum_depivot(month_trailing_df, popularity_df)
semiannual_depivot = momentum_depivot(semiannual_trailing_df, popularity_df)
annual_depivot     = momentum_depivot(annual_trailing_df, popularity_df)
```

```python
def plot_series(ticker, trailing):
    df = {
        'quarter': quarter_depivot,
        'month': month_depivot,
        'semiannual': semiannual_depivot,
        'annual': annual_depivot
    }[trailing]
    
    df[df.Ticker == ticker].set_index('Date').plot(figsize=(8, 8))
    plt.xlabel('Date')
    plt.ylabel('Rank')
plot_series('ZUMZ', 'month')
```
